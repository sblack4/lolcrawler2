```python
import pandas as pd 
```

```python
df = pd.read_csv("../data/GoldDiffBinned.csv")
```

```python
df.head()
```

```python
df = df.drop(['ind', 'winnerTotal', 'loserTotal'], axis='columns')
df.head()
```

```python
df_trans = pd.pivot_table(df, values='GoldDifference', index='gameID', columns='bin')
df_trans.head()
```

```python
df_trans.reset_index(level=0, inplace=True)
df_trans.head()
```

```python
df_trans.to_csv("../data/GoldDiffTrans.csv")
```
